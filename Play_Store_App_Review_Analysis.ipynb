```python
#Function to convert Mb to Kb

def mb_kb_convertor(text):
    # Check if the text ends with 'M' (Megabytes)
    if text.endswith('M'):
        # Convert the value to kilobytes by multiplying by 1024
        return float(text[:-1]) * 1024
    # Check if the text ends with 'k' (Kilobytes)
    elif text.endswith('k'):
        # Directly convert the value to float (already in kilobytes)
        return float(text[:-1])
    else:
        # Return the text as is if it does not end with 'M' or 'k'
        return text

# Apply the function to the 'Size' column
df1['Size'] = df1['Size'].apply(mb_kb_convertor)


df1['Size']
```

**Feature engineering:**

Create a new feature Size Category from Size column

```python
# Define a function to categorize sizes into 10 categories
def categorize_size(size):
    if size == "Varies with device":
        return "Varies with device"
    elif size < 100:  # less than 0.1MB
        return "Less than 0.1MB"
    elif 100 <= size < 1024:  # 0.5MB to 1MB
        return "0.1MB to 1MB"
    elif 1024 <= size < 5120:  # 1MB to 5MB
        return "1MB to 5MB"
    elif 5120 <= size < 10240:  # 5MB to 10MB
        return "5MB to 10MB"
    elif 10240 <= size < 20480:  # 10MB to 20MB
        return "10MB to 20MB"
    elif 20480 <= size < 51200:  # 20MB to 50MB
        return "20MB to 50MB"
    elif 51200 <= size < 102400:  # 50MB to 100MB
        return "50MB to 100MB"
    else:  # greater than 100MB
        return "Greater than 100MB"

# Apply the function to create a new column 'Size Category'
df1['Size Category'] = df1['Size'].apply(categorize_size)


# Optionally, check the distribution of the size categories
df1['Size Category'].value_counts()
```

# **identifiy outliers in our data:**
**Common Approaches to Detect Outliers:**

**Visual Inspection:**

   **Boxplots:** Effective for visualizing the distribution of data and identifying outliers.

  **Scatter Plots:** Useful for identifying outliers in bivariate data.

**Statistical Methods:**

**Z-Score:** Measures how many standard deviations a data point is from the mean.

**IQR Method:** Identifies outliers based on the interquartile range (IQR).

```python
#Using Boxplot
#Boxplots are a straightforward way to visualize outliers in numerical data.

app=cleaned_app.copy()

# Define numerical columns
numerical_columns = app.select_dtypes(include=['int','float']).columns

# Plot boxplots for numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(data=app[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

**Insights from Box Plot:**

Rating: The high median and narrow IQR indicate that most apps are well-rated.

Reviews and Installs: Both columns show high variability and a large number of outliers, suggesting a small number of apps are much more popular than the majority.

Price: Most apps are free, but there are some high-priced outliers.

These insights can guide further analysis, such as investigating the characteristics of apps with high reviews and installs or understanding the impact of pricing on app popularity.

```python
#Using Z_Score
#The Z-score can help identify outliers by measuring how far away a data point is from the mean.

from scipy import stats

# Calculate Z-scores for numerical columns
z_scores = stats.zscore(app[numerical_columns].dropna())

# Identify outliers (abs(Z-score) > 3 is a common threshold)
outliers = (abs(z_scores) > 3).any(axis=1)

# Reset index of outliers to align with the original DataFrame 'app'
outliers = outliers.reindex(app.index, fill_value=False)

# Display outliers
app_outliers = app[outliers]
app_outliers
```

```python
#Using IQR Method: This method is effective for detecting outliers by considering values beyond the typical range.

# Identify outliers using IQR method
def find_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))

# Apply the IQR method to numerical columns
outliers_iqr = app[numerical_columns].apply(find_outliers_iqr)

# Display rows with outliers
app_outliers_iqr = app[(outliers_iqr.any(axis=1))]
app_outliers_iqr
```

# Reason why we should not remove outliers during exploratory data analysis (EDA):
1. **Rating Column**

Example Scenario:

Apps in the dataset have ratings ranging from 2.0 to 4.8.

An outlier might be an app with a rating of 1.0 or 5.0.

**Reasons Not to Remove Outliers:**

Representative of Extreme User Feedback: Outliers like a rating of 1.0 or 5.0 reflect extreme user experiences that are critical for understanding user sentiment towards apps.
Impact on App Evaluation: Removing extreme ratings could distort the analysis of app quality metrics, such as average ratings across categories or genres.
Insight into User Preferences: Extreme ratings provide insights into highly polarizing apps, helping to identify which apps resonate strongly with users and which do not.

**2. Reviews Column**

Example Scenario:

App A has 1,000,000 user reviews.

App B has 500,000 user reviews.

App C has 10 user reviews (potential outlier).

**Reasons Not to Remove Outliers:**

User Engagement and Popularity: App C with significantly fewer reviews may indicate a new or niche app that hasn't yet gained widespread attention but could be emerging in popularity.

Understanding User Feedback: Analyzing why certain apps have lower review counts offers insights into user engagement patterns and app usability.

Market Positioning: Retaining outliers helps in understanding the market position of apps with varying levels of user interaction, essential for
competitive analysis and strategic decision-making.

**3. Installs Column**

Example Scenario:

App X has 10,000,000 installs.

App Y has 5,000,000 installs.

App Z has 100 installs (potential outlier).

**Reasons Not to Remove Outliers:**

Market Segmentation: App Z with fewer installs may target a specific niche market or demographic that is underserved by more popular apps.

Emerging Trends: Identifying apps with lower install counts can reveal emerging trends or new entrants in the app market that have potential for growth.

Strategic Insights: Retaining outliers provides insights into market dynamics and consumer behavior, supporting innovation and market expansion strategies.

**4. Price Column**

Example Scenario:

Most apps are free (Price = 0).

Some apps are paid with prices ranging from 0.99 to 9.99.

An outlier might be an app priced at $99.99.

**Reasons Not to Remove Outliers:**

Revenue and Monetization: High-priced apps may cater to niche markets willing to pay premium prices for specialized features or content, contributing significantly to revenue generation.

Understanding Pricing Strategies: Analyzing outlier prices offers insights into app monetization models and competitive pricing strategies.

Market Dynamics: Retaining outliers ensures a comprehensive view of the competitive landscape, especially in markets where pricing diversity is a strategic advantage.

# **Temporal Trends**
**Time Series Analysis:** Analyze how app updates (Last Updated) correlate with ratings, reviews, and installs over time. Use line plots or scatter plots with a time axis to visualize trends.

**Seasonality:** Identify any seasonal patterns in app updates or user engagement metrics that could influence app performance.

**Purpose:** Identify patterns over time, seasonality, and trends.

# Correlation between Year of last Update and Rating.

```python
# Group by year and calculate mean rating
average_ratings_by_year = app.groupby([i.year for i in list(app["Last Updated"])])["Rating"].mean()

# Extract years and corresponding average ratings
years = list(set([i.year for i in list(app["Last Updated"])]))
sorted_years = sorted(years)
average_ratings = list(average_ratings_by_year)

# Plot the data
plt.plot(sorted_years, average_ratings, "o-g")
plt.title("Last year vs Average rating")
plt.xlabel("Year of last update")
plt.ylabel("Average rating")
plt.show()
```

# From this line plot we have concluded that:

Average app rating is gradually increasing after 2016. That means apps that have regular updates have higher average ratings.
The graph is falling from 2010 to 2012. This may be due to bad user experience, people are started giving poor ratings due to non regular updates or many other reasons too.

# Correlation between Year of last Update and Reviews.

```python
# "Last Updated" column is in datetime format
app['Last Updated'] = pd.to_datetime(app['Last Updated'])

# Group by year and calculate mean reviews
avg_reviews_by_year = app.groupby(app['Last Updated'].dt.year)['Reviews'].mean()

# Extract years and corresponding average reviews
years = sorted(avg_reviews_by_year.index)
average_reviews = avg_reviews_by_year.values

# Plot the data
plt.plot(years, average_reviews, "o-g")
plt.title("Last year vs Average Reviews")
plt.xlabel("Year of last update")
plt.ylabel("Average Reviews")
plt.show()
```

# Insights:
**Steady Growth (2010-2013):** There is a clear upward trend in average reviews, indicating increasing user engagement with apps during this period.

**Fluctuations (2014-2017)**: Average reviews show high but fluctuating values, suggesting stable yet variable user interactions.

**Significant Spike (2018):** A dramatic rise in 2018 with average reviews reaching approximately 314,807, potentially due to major market events or increased app popularity.

# **New Section**
Correlation between Year of last Update and Avg Installs.

```python
# Group by year and calculate mean installs
average_installs_by_year = app.groupby([i.year for i in list(app["Last Updated"])])["Installs"].mean()

# Extract years and corresponding average installs
years = list(set([i.year for i in list(app["Last Updated"])]))
sorted_years = sorted(years)
average_installs = [average_installs_by_year[year] for year in sorted_years]

# Plot the data
plt.plot(sorted_years, average_installs, "o-g")
plt.title("Last year vs Average installs")
plt.xlabel("Year of last update")
plt.ylabel("Average installs")
plt.show()
```

# Inference
**Significant Increase in 2018**: Average installs dramatically increased to approximately 11.18 million, indicating a major surge in app popularity.

**Steady Growth with Fluctuations**: From 2010 to 2018, average installs showed a steady increase with some fluctuations, highlighting periods of high user engagement.

**Early Market Growth (2010-2011):** A sharp rise from 100,000 installs in 2010 to over 1 million in 2011 suggests rapid growth in the early app market.

Overall, the data shows growing user engagement and app adoption, with a notable peak in 2018.
