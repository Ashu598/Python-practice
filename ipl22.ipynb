```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
ipl = pd.read_csv("ipl22.csv")
ipl.shape
ipl.head(3)
```

```python
ipl["player_of_the_match"].value_counts()  #will groupby and count
a = ipl["player_of_the_match"].value_counts() [0:4]     #will result top 4 values
a
```

```python
#Create a bar of top 8 player of the match
plt.figure(figsize =(10,3))
plt.bar(list(ipl["player_of_the_match"].value_counts().keys())[:8], list(ipl["player_of_the_match"].value_counts()[:8]) , color = 'g')
plt.show()
```

```python
#Find top performer teams in descending order
ipl['match_winner'].value_counts()
```

```python
#Make a plot chart of top performer teams in 2022
plt.figure(figsize= (10,4))
plt.plot(list(ipl['match_winner'].value_counts().keys()), list(ipl['match_winner'].value_counts()))
plt.show()
```

```python
#find the list of top 5 players having highest scores
c= ipl.nlargest(5, ['highscore'])    #used pandas nlargest
c
```

```python
#make a scatter graph that shows players with highscores
plt.figure(figsize = (8,4))
plt.scatter(c['top_scorer'], c['highscore'], color="y")
plt.title("Top scorers in IPL 2022")
plt.xlabel("Players Name")
plt.ylabel("Players Scores")
plt.show()
```
